```python
# Gerekli ktphaneleri ykleme
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veriyi ykleme (rnek olarak bir CSV dosyas kullanyoruz)
url = 'https://raw.githubusercontent.com/datasets/covid-19/master/data/time-series-19-covid-combined.csv'
df = pd.read_csv(url)

# Veriyi inceleme
print(df.head())

# Basit bir grselletirme rnei
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Confirmed', hue='Country/Region')
plt.title('lkeler Baznda COVID-19 Vaka Saylar')
plt.xlabel('Tarih')
plt.ylabel('Vaka Says')
plt.xticks(rotation=45)
plt.show()
```
