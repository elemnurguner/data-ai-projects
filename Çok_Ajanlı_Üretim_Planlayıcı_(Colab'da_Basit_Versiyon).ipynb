```python
# 1. GEREKLİ KURULUMLAR (Önce bu hücreyi çalıştırın)
!pip install -q llama-cpp-python langchain crewai==0.28.8

# 2. DAHA KÜÇÜK MODEL İNDİRME (Daha az bellek kullanır)
!wget -O model.gguf https://huggingface.co/TheBloke/phi-2-GGUF/resolve/main/phi-2.Q4_K_M.gguf

# 3. KÜTÜPHANELER
from langchain.llms import LlamaCpp
from crewai import Agent, Task, Crew

# 4. OPTİMİZE EDİLMİŞ MODEL YÜKLEME
llm = LlamaCpp(
    model_path="model.gguf",
    temperature=0.3,
    max_tokens=768,  # Daha kısa çıktılar için
    n_ctx=2048,
    n_gpu_layers=20,  # GPU katman sayısını azalttık
    n_batch=256,      # Batch boyutunu küçülttük
    verbose=False
)

# 5. OPTİMİZE EDİLMİŞ AJANLAR
production_manager = Agent(
    role="Üretim Müdürü",
    goal="Montaj hattı verimliliğini maksimize et",
    backstory="20 yıllık otomasyon uzmanı",
    llm=llm,
    allow_delegation=False,
    verbose=True,
    max_iter=3  # Maksimum iterasyon sayısını sınırla
)

quality_agent = Agent(
    role="Kalite Müfettişi",
    goal="Ürün kalitesini garanti altına al",
    backstory="Metroloji alanında uzman",
    llm=llm,
    allow_delegation=False,
    verbose=True,
    max_iter=3
)

# 6. DAHA KISA GÖREVLER
tasks = [
    Task(
        description="""Verimlilik analizi yap:
        Makine1: %78
        Makine2: %92
        Makine3: %65
        Darboğazı ve 2 çözüm öner""",
        agent=production_manager,
        expected_output="2 maddelik çözüm",
        async_execution=False,
        output_file="verimlilik_raporu.txt"
    ),
    Task(
        description="""Kalite analizi yap:
        Makine1: %78
        Makine2: %92
        Makine3: %65
        2 öneri ver""",
        agent=quality_agent,
        expected_output="2 maddelik öneri",
        async_execution=False,
        output_file="kalite_raporu.txt"
    )
]

# 7. EKİP OLUŞTURMA
crew = Crew(
    agents=[production_manager, quality_agent],
    tasks=tasks,
    verbose=2,
    memory=False  # Hafızayı kapatarak kaynak tüketimini azalt
)

# 8. ÇALIŞTIRMA
try:
    result = crew.kickoff(inputs={"timeout": 180})  # 3 dakika timeout
    print("\n⚙️ ÜRETİM ANALİZ SONUÇLARI:")
    print(result)
except Exception as e:
    print(f"⛔ Hata oluştu: {str(e)}")
    print("ℹ️ Çözüm önerileri:")
    print("- Colab Runtime'ı yeniden başlatın")
    print("- Daha basit görevler deneyin")
    print("- !pip install --upgrade llama-cpp-python")
```
