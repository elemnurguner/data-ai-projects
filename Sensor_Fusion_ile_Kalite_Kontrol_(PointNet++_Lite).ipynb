```python
!pip install open3d numpy
import open3d as o3d
import numpy as np

# Basit nokta bulutu oluşturma
points = np.random.rand(1000, 3) * 10  # 1000 rastgele nokta
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(points)

# Basit kümeleme analizi
labels = np.array(pcd.cluster_dbscan(eps=0.5, min_points=10))
max_label = labels.max()
colors = np.random.rand(max_label + 1, 3)
colors = [colors[label] if label >= 0 else [0,0,0] for label in labels]
pcd.colors = o3d.utility.Vector3dVector(colors[:len(points)])

o3d.visualization.draw_plotly([pcd])  # Colab'da görselleştirme
```
